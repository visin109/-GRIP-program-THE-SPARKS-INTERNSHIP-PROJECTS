# Now let's train the machine using a classification algorithm called DECISION TREE CLASSIFIER which 
# mostly works based on many if or else conditions

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
X=iris.drop('Species',axis=1)
y=iris['Species']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train
y_train
```

# Predicting the model and converting into dataframe by testing it on one of the features(SepalWidth here)

```python
model=DecisionTreeClassifier(max_depth=4).fit(X_train,y_train)
model
predict=model.predict(X_test)
pd.DataFrame(predict,index=X_test['SepalWidth'],columns={'Species'})
```

# Visualizing the working of  decision tree 

```python
features=['SepalLength','SepalWidth','PetalLength','PetalWidth']
target=['setosa','versicolor','virginica']
plt.figure(figsize=(20,20))
plot_tree(model,feature_names=features,class_names=target,filled=True)
```
